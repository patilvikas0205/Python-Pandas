### Step 3. Assign it to a variable called users.

```python
users=pd.read_csv("u.user",sep="|")
users
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(by="occupation").mean()
```

```python
users.groupby(by="occupation").mean()["age"]
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
'''
# create a function
def gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0
'''
# apply the function to the gender column and create a new column
users['gender_n'] = [1 if i=="M" else 0 for i in users.gender]


(users.groupby('occupation').gender_n.sum() / users.occupation.value_counts() * 100).sort_values(ascending = False)

# sort to the most male 
#a.sort_values(ascending = False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby(by="occupation").agg([min,max])["age"]
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(by=["occupation","gender"]).mean()["age"]
```

### Step 8.  For each occupation present the percentage of women and men

```python
a=users.groupby(by=["occupation"]).count()
print(a)
```

```python
g=users.groupby(by=["occupation","gender"]).agg({'gender': 'count'})
print(g)
```
