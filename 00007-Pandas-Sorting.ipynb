# pandas sorting


```python
import pandas as pd
```

```python
data=[["A",40,50,60,30],["B",85,45,79,12],["C",42,15,60,24],
      ["D",25,78,45,62],["E",45,85,24,56],["F",52,48,56,74],
      ["G",56,87,45,58],["H",56,48,89,52],["I",58,45,66,55],
      ["J",56,87,25,45]]
```

```python
df1=pd.DataFrame(data=data,columns=["Name","Maths","Physics","Chemistry","English"])
```

```python
df1
```

# Sort by Label

```python
#Sort by index
df1.sort_index()
```

```python
#sort by axis default axis value is 0 row wise
df1.sort_index()
```

```python
#sort by axis 1 Column wise
df1.sort_index(axis=1)
```

```python
#sort by descending
#by default is ascending 
df1.sort_index(ascending=False)
```

```python
#same as sort by axis 1 ie by columns
df1.sort_index(1)
```

```python
#sort by values pass column name by default is ascending
df1.sort_values("Maths")
```

```python
#sort by values oerder pass column name by default is ascending ==> descending
df1.sort_values("Maths", ascending=False)
```

```python
#sort by algorithm  ==> sort_values(by=Column_name, kind="sort_name")
df1.sort_values(by="Maths",kind="mergesort")
```

```python
#sort by algorithm  ==> sort_values(by=Column_name, kind="sort_name")
df1.sort_values(by="Physics",kind="quicksort")
```

```python
#sort by algorithm  ==> sort_values(by=Column_name, kind="sort_name") in descending oerder
df1.sort_values(by="Physics",kind="quicksort",ascending=False)
```
